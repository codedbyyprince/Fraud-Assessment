```python
def make_env_risk(freq):
    if freq < 0.005:
        return 2      # very rare
    elif freq < 0.05:
        return 1      # rare
    else:
        return 0      # common

identity['environment_risk'] = identity['environment_freq'].apply(make_env_risk)
```

```python
def device_os_invalid(device, os):

    if device == 'NA' or os == 'NA' or device == 'Other' or os == 'Other':
        return 0

    if device == 'mobile':
        if os in ['Android', 'iOS']:
            return 0
        else:
            return 1

    if device == 'desktop':
        if os in ['Windows', 'macOS', 'Linux']:
            return 0
        else:
            return 1

    return 0
```

```python
identity['device_os_invalid'] = identity.apply(
    lambda row: device_os_invalid(row['DeviceType'], row['user_os']),
    axis=1
)
```
