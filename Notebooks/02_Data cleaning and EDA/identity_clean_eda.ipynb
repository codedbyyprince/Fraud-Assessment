## **02_Identity EDA**

### **Univariate**

```python
identity['DeviceType'] = identity['DeviceType'].fillna('NA')
counts = identity['DeviceType'].value_counts()

plt.bar(counts.index, counts.values)
plt.xlabel("Device Type")
plt.ylabel("Counts")
plt.title('DeviceType Distribution (Including NA)')
plt.show()
```

```python
identity['is_identity_seen_before'] = identity['is_identity_seen_before'].fillna('NA')
counts = identity['is_identity_seen_before'].value_counts()

plt.bar(counts.index, counts.values)
plt.xlabel("is identity seen before")
plt.ylabel("Counts")
plt.title('identity seen before Distribution (Including NA)')
plt.show()
```

```python
identity['user_os'] = identity['user_os'].fillna('NA')
counts = identity['user_os'].value_counts()

plt.bar(counts.index, counts.values)
plt.xlabel("user_os")
plt.ylabel("Counts")
plt.title('user_os Distribution (Including NA)')
plt.show()
```

```python
identity['user_browser'] = identity['user_browser'].fillna('NA')
counts = identity['user_browser'].value_counts()

plt.figure(figsize=(10, 6))
plt.barh(counts.index, counts.values)

plt.xlabel("Counts")
plt.ylabel("User Browser")
plt.title("User Browser Distribution (Including NA)")

plt.tight_layout()
plt.show()
```
